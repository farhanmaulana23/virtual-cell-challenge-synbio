```python
from pathlib import Path
import scanpy as sc
import pandas as pd

#paths
DATA_DIR = Path("data")
train_file = DATA_DIR / "adata_Training.h5ad"
genes_file = DATA_DIR / "gene_names.csv"

#load training set
adata = sc.read_h5ad(train_file, backed="r")
print("Shape:", adata.shape)               # (cells, genes)
print("Obs columns:", adata.obs.columns[:6])
print("Var (first 5 genes):", adata.var_names[:5])

#basic QC metrics
adata.obs['n_counts'] = adata.X.sum(1)
adata.obs['n_genes']  = (adata.X > 0).sum(1)

sc.pl.violin(
    adata,
    ['n_counts', 'n_genes'],
    log=True,
    stripplot=False,
    multi_panel=True
)

#verify gene order vs. challenge
gene_list = pd.read_csv(genes_file, header=None)[0].tolist()
print("Genes in challenge list:", len(gene_list))
print("Overlap with AnnData:", len(set(gene_list) & set(adata.var_names)))
```

```python
import numpy as np
import scanpy as sc
from pathlib import Path

DATA_DIR   = Path("data")
train_path = DATA_DIR / "adata_Training.h5ad"

# open backed
adata_backed = sc.read_h5ad(train_path, backed="r")

# take a 10 k-cell random subset
idx = np.random.choice(adata_backed.shape[0], size=10_000, replace=False)
adata = adata_backed[idx].to_memory()      # materialise that slice in RAM

# all dense/sparse ops work
adata.obs['n_counts'] = adata.X.sum(axis=1).A1        # .A1 flattens to 1-D
adata.obs['n_genes']  = (adata.X > 0).sum(axis=1).A1

sc.pl.violin(
    adata,
    ['n_counts', 'n_genes'],
    log=True,
    stripplot=False,
    multi_panel=True
)
```
